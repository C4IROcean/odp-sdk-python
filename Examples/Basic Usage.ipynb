```python

from odp_sdk import client
client=ODPClient('xxxxxxxxxxxxxx') #Connect to the Ocean Data Platform
```

### Download Ocean Data #1 (simple)

```python
df1=client.casts(longitude=[-25,35],
                latitude=[50,80],
                timespan=['2018-06-01','2018-06-30']) #Download Ocean Data to pandas dataframe
df1.head() # display first rows
```

### Download Ocean Data #2 (with more options) 

```python
df2=client.casts(longitude=[-25,35],
                latitude=[50,80],
                timespan=['2018-06-01','2018-06-30'],
                parameters=['date','lon','lat','z','Temperature','Salinity'],
                n_threads=35,
                include_flagged_data=False) 
df2.head()
```

### Download Ocean Data #3 (detailed step by step)

```python
casts=client.get_available_casts(2017,2019) # Download an overview of the available casts from selected years
casts.head()
```

```python
casts_filtered=client.filter_casts(casts,
                                   longitude=[-25,35],
                                   latitude=[50,80],
                                   timespan=['2018-06-01','2018-06-30'])
casts_filtered.head()
```

```python
df3=client.download_data_from_casts(casts_filtered.extId.tolist(),
                                n_threads=35,
                                parameters=['date','lon','lat','z','Temperature','Salinity'])
df3.head()
```
